```python
import csv

def add_user():
    new_username = input("What's your username? ").strip()
    new_password = input("What's your password? ").strip()
    new_admin = input("Are they admin? (yes/no): ").lower().strip()

    with open("database.csv", "r") as file:
        reader = csv.reader(file)
        for user in reader:
            if user[0] == new_username:      #checking if username is taken
                print("This username is already taken.")
                return

    with open("database.csv", "a") as file:
        #writer = csv.writer(file)
        #writer.writerow([new_username, new_password, new_admin])
        file.write(f"{new_username}, {new_password}, {new_admin}\n")
        print(f"User {new_username} added successfully!")
#---------------------------------------------------------------------------

def see_users():
    with open("database.csv") as file:
        reader = csv.reader(file)
        for user in reader:
            print(f"Username: {user[0]}")

#----------------------------------------------------------------------------

def delete_user():
    username_to_delete = input("Enter the username you want to delete: ").strip()

    with open("database.csv", "r") as file:
        reader = csv.reader(file)
        users = list(reader)

    user_found = False
    with open("database.csv", "w", newline='') as file:
        writer = csv.writer(file)
        for user in users:
            if user[0] == username_to_delete:
                print(f"User {username_to_delete} found and deleted.")
                user_found = True
                continue  # Skip this user to delete them
            writer.writerow(user)  # Write other users back

    if not user_found:
        print(f"User {username_to_delete} not found in the database.")

#----------------------------------------------------------------------------

def update_user():
    username_to_update = input("Enter the username you want to update: ").strip()

    # Get new details for the user to update
    new_username = input("Enter the new username: ").strip()
    new_password = input("Enter the new password: ").strip()
    new_admin = input("Is the user admin? (yes/no): ").strip().lower()

    with open("database.csv", "r") as file:
        reader = csv.reader(file)
        for user in reader:
            if user[0] == new_username:   #check if the username is already used
                print("This username is already taken.")
                return    #stops here if the username is used

    # Read all users from the file into memory
    with open("database.csv", "r") as file:
        reader = csv.reader(file)
        users = list(reader)  # Read all rows and store them in a list

    # Flag to check if the user was found
    user_found = False

    # Open the file in write mode to overwrite with updated data
    with open("database.csv", "w", newline='') as file:
        writer = csv.writer(file)
        
        for user in users:
            if user[0] == username_to_update:
                # Update the user details
                print(f"User {username_to_update} found and updated.")
                writer.writerow([new_username, new_password, new_admin])
                user_found = True
            else:
                # Write the unchanged users back to the file
                writer.writerow(user)

    # If the user wasn't found, notify the user
    if not user_found:
        print(f"User {username_to_update} not found in the database.")

#----------------------------------------------------------------------------

def main():
    try:
        user_username = input("Enter your username: ").strip()
        user_password = input("Enter your password: ").strip()

        with open("database.csv") as file:
            reader = csv.reader(file)
            for row in reader:
                username = row[0].strip()      
                password = row[1].strip()
                admin = row[2].strip().lower()

                if user_username == username and user_password == password:
                    if admin == "yes":       #here it will give the admin there meny
                        while True:
                            print("-------ADMIN-------")
                            print("\nPlease make a selection:")
                            print("1 - View users")
                            print("2 - Add users")
                            print("3 - Delete users")
                            print("4 - Update users")
                            print("5 - Done")

                            admin_input = input("Enter your choice: ")
                            if admin_input == "5":
                                break
                            elif admin_input == "1":
                                see_users()
                            elif admin_input == "2":
                                add_user()
                            elif admin_input == "3":
                                delete_user()
                            elif admin_input == "4":
                                update_user()
                            else:
                                print("Invalid choice. Please try again.")
                    else:
                        print("\nNon-admin users:")
                        see_users()
                    break
            else:
                print("Invalid username or password.")

    except FileNotFoundError:
        print("File not found. Please check the file name.")
    except ValueError:
        print("Invalid data in file.")


if __name__ == "__main__": 
    main()
```
